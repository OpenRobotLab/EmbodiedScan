## Data Visualizations

We provide list and visualize methods. They are implemented in the EmbodiedScanExplorer class.

EmbodiedScan is composed of three dataset sources. Refer to `install.md` to prepare raw data.

EmbodiedScanExplorer can list and visualize existing scenes and images if some composed dataset are missing.

```python
from embodiedscan.explorer import EmbodiedScanExplorer

# If there is only one composed dataset to explore
# explorer = EmbodiedScanExplorer(
# 	dataroot='path/to/dataset',
# 	ann_file='path/to/annotation.pkl',
# 	verbose=True,	# print log or not
# )

# # If there are multiple composed dataset and multiple annotation files
# # this requires 'path/to/dataset' contain the name of composed dataset.
# explorer = EmbodiedScanExplorer(
# 	dataroot=['path/to/scannet', 'path/to/3rscan', 'path/to/matterport3d'],
# 	ann_file=['path/to/train.pkl', 'path/to/val.pkl'],
# 	verbose=True,	# print log or not
# )

# or
explorer = EmbodiedScanExplorer(
	data_root={'scannet' : 'demo/data/scannet', '3rscan' : 'demo/data/3rscan', 'matterport3d': 'demo/data/matterport3d'},
	ann_file=['demo/data/train.pkl', 'demo/data/val.pkl'],
	verbose=True,	# print log or not
)
```

### List methods
There are four list methods available.

1. `list_categories()` lists all categories.

```python
explorer.list_categories()
```

2. `list_scenes()` lists all scenes in the loaded DB.

```python
explorer.list_scenes()
```

3. `list_cameras()` lists all cameras in a given scene.

```python
explorer.list_cameras(scene='3rscan/02b33dfb-be2b-2d54-92d2-cd012b2b3c40')
```

4. `list_instances()` lists all annotated instances in a given scene.

```python
explorer.list_instances(scene='3rscan/02b33dfb-be2b-2d54-92d2-cd012b2b3c40')
```

### Visualize methods
There are three visualize methods available.

1. `show_image()` shows the captured image by a given camera in a given scene. Boxes will be drawed into the image if `render_box=True`

```python
explorer.show_image('scannet/scene0000_00', '00000', render_box=True)
```

2. `render_scene()` renders the mesh of a given scene. This function will create a new window to show the results.

```python
explorer.render_scene('3rscan/02b33dfb-be2b-2d54-92d2-cd012b2b3c40', render_box=True)
```

3. `render_occupancy()` renders the occupancy voxels of a given scene. This function will create a new window to show the results.

```python
explorer.render_occupancy('scannet/scene0000_00')
```

4. `render_continuous_scene()` renders the continuous frames of a given scene. This function will create a new window to show the results.

```python
explorer.render_continuous_scene('scannet/scene0000_00', start_cam='00000', pcd_downsample = 100)	# set a large pcd_downsample if you want to render a lot of frames
```

5. `render_continuous_occupancy()` renders the continuously updated occupancy of a given scene. This function will create a new window to show the results.

```python
explorer.render_continuous_occupancy('scannet/scene0000_00', start_cam='00000')
```
